```python
import numpy as np
```

```python
def gambler_value_iteration(goal=100, p_win=0.4, theta=1e-9):
    V = np.zeros(goal + 1)
    Policy = np.zeros(goal + 1, dtype=int)
    V[goal] = 1

    while True:
        delta = 0
        for s in range(1, goal):
            best_value = 0
            best_action = 0
            for a in range(1, min(s, goal - s) + 1):
                win_state = s + a
                lose_state = s - a
                value = p_win * V[win_state] + (1 - p_win) * V[lose_state]
                if value > best_value:
                    best_value = value
                    best_action = a
            delta = max(delta, abs(best_value - V[s]))
            V[s] = best_value
            Policy[s] = best_action if best_value > 0 else 1
        if delta < theta:
            break

    return Policy, V
```

```python
def print_optimal_policy(policy, goal):
    """Prints the optimal betting strategy"""
    print("Optimal Betting Policy:")
    for s in range(1, goal):
        print(f"Capital {s}: Bet {policy[s]}")
```

```python
if __name__ == "__main__":
    goal_amount = 100
    win_probability = 0.4
    policy, values = gambler_value_iteration(goal=goal_amount, p_win=win_probability)
    print_optimal_policy(policy, goal_amount)
```
