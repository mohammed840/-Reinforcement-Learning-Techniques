```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_steps = 1000
start_position = 0
transitions = [-1, 1]
probabilities = [0.5, 0.5]
random_walk = [start_position]
```

```python
for _ in range(n_steps):
    step = np.random.choice(transitions, p=probabilities)
    random_walk.append(random_walk[-1] + step)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(random_walk)
plt.title('Random walk using Markov Process')
plt.xlabel('Step')
plt.ylabel('Position')
plt.grid(True)
plt.show()
```
